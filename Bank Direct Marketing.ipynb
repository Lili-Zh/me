### Problem 10: Model Comparisons

Now, we aim to compare the performance of the Logistic Regression model to our KNN algorithm, Decision Tree, and SVM models.  Using the default settings for each of the models, fit and score each.  Also, be sure to compare the fit time of each of the models.  Present your findings in a `DataFrame` similar to that below:

| Model | Train Time | Train Accuracy | Test Accuracy |
| ----- | ---------- | -------------  | -----------   |
|     |    |.     |.     |

````python
import pandas as pd
import time
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

df = pd.read_csv(r'C:\Users\liliz\OneDrive\Desktop\AI and ML\Berkeley Certificate\Module 17\module17_starter\data\bank-additional-full.csv', sep = ';')

# 2. Drop 'duration' (known data leak)
if 'duration' in df.columns:
    df.drop(columns='duration', inplace=True)

# 3. Keep only valid 'y' values: 'yes' or 'no'
df = df[df['y'].isin(['yes', 'no'])]

# 4. Map target variable to binary: 'yes' → 1, 'no' → 0
df['y'] = df['y'].map({'yes': 1, 'no': 0})

# 5. One-hot encode categorical variables
categorical_cols = df.select_dtypes(include='object').columns.tolist()
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# 6. Split into features and target
X = df_encoded.drop('y', axis=1)
y = df_encoded['y']

# 7. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 8. Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 9. Define models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'KNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'SVM': SVC()
}

# 10. Train, time, and evaluate models
results = []
for name, model in models.items():
    start = time.time()
    model.fit(X_train_scaled, y_train)
    end = time.time()

    train_pred = model.predict(X_train_scaled)
    test_pred = model.predict(X_test_scaled)

    train_acc = accuracy_score(y_train, train_pred)
    test_acc = accuracy_score(y_test, test_pred)
    train_time = round(end - start, 4)

    results.append({
        'Model': name,
        'Train Time': train_time,
        'Train Accuracy': round(train_acc, 4),
        'Test Accuracy': round(test_acc, 4)
    })

# 11. Show results
results_df = pd.DataFrame(results)
print(results_df)
````

````python
from sklearn.model_selection import GridSearchCV

# KNN: Tune number of neighbors
knn_grid = {'n_neighbors': [3, 5, 7, 9]}
knn_search = GridSearchCV(KNeighborsClassifier(), knn_grid, cv=5, scoring='accuracy')
knn_search.fit(X_train_scaled, y_train)

# Decision Tree: Tune max depth and min samples
tree_grid = {
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10]
}
tree_search = GridSearchCV(DecisionTreeClassifier(random_state=42), tree_grid, cv=5, scoring='accuracy')
tree_search.fit(X_train_scaled, y_train)

# SVM: Tune kernel and regularization
svm_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
svm_search = GridSearchCV(SVC(), svm_grid, cv=5, scoring='accuracy')
svm_search.fit(X_train_scaled, y_train)

# Print best hyperparameters and scores
print("Best KNN:", knn_search.best_params_, "| Accuracy:", knn_search.best_score_)
print("Best Decision Tree:", tree_search.best_params_, "| Accuracy:", tree_search.best_score_)
print("Best SVM:", svm_search.best_params_, "| Accuracy:", svm_search.best_score_)
````

````python
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

# Evaluate best KNN model
y_pred = knn_search.predict(X_test_scaled)
y_proba = knn_search.predict_proba(X_test_scaled)[:, 1]

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
````
